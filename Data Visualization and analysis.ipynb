# 2) Data Visualization

```python
filt = vac_data.country=='China'
vac_data.loc[filt]
```

    Countrywise data

```python
country_grp = vac_data.groupby(['country'])
```

```python
country_grp['total_vaccinations']
```

```python
con_wise_df = country_grp.agg({'total_vaccinations': 'max',
                 'people_vaccinated_only_1st_dose':'max',
                 'people_fully_vaccinated':'max',
                 'daily_vaccinations_avgd':'median',
                 '%_people_vaccinated':'max'})
```

```python
con_wise_df
```

```python
con_wise_df
```

```python
con_wise_total_vac = con_wise_df['total_vaccinations'].sort_values(ascending=False).head(20)
con_wise_total_vac
```

```python
con_wise_total_vac.sort_values(ascending=True, inplace=True)
```

```python
# con_wise_total_vac.plot.bar(figsize=(10,6))

plt.barh(con_wise_total_vac.index, con_wise_total_vac.values)
plt.xlabel('Conutry')
plt.ylabel('Total Vaccinations')
plt.title('Top 20 Countries with Maximum Total Vaccinations')
```
