```python
prev_state=vac_df.index[0][0]
prev_vac=0
curr_vac=0
count = 0

for state, date in vac_df.index:
    curr_state = state
    
    if(curr_state!=prev_state): 
        prev_vac=0
        
    curr_vac = vac_df.loc[(state,date),'Total Individuals Vaccinated']
    if(curr_vac-prev_vac < 0 ): 
        count+=1;
        vac_df.loc[(state,date), 'Daily Individuals Vaccinated'] = 0
    else: vac_df.loc[(state,date), 'Daily Individuals Vaccinated'] = curr_vac-prev_vac
    prev_vac=curr_vac
    prev_state = curr_state
```
