```python
vac_df = pd.read_csv('cowin_vaccine_data_statewise.csv')
pd.set_option('display.max_rows', 100)
```

```python
filt = vac_df.State=='India'
ind_vac_df = vac_df.loc[filt].copy()
```

```python
ind_vac_df
```

```python
ind_vac_df.dropna(inplace=True)
```

```python
ind_vac_df
```

```python
daily_vac = []
```

```python
daily_vac
```

```python
ind_vac_df.shape[0]
```

```python
ind_vac_df['Total Individuals Vaccinated'][0]
```

```python
prev_vac=0;
curr_vac=0;

for i in range(ind_vac_df.shape[0]):
    curr_vac=ind_vac_df['Total Individuals Vaccinated'][i]
    daily_vac.append(curr_vac - prev_vac)
    prev_vac=curr_vac
```

```python
ind_vac_df['daily Individuals Vaccinated'] = daily_vac
```
