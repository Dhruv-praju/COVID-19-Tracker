# 2) Data Visualization

```python
plt.style.use('seaborn')
filt = vac_data.country=='Gibralter'
vac_data.loc[filt]
```

    Countrywise data

```python
country_grp = vac_data.groupby(['country'])
```

```python
country_grp['total_vaccinations']
```

```python
con_wise_df = country_grp.agg({'total_vaccinations': 'max',
                 'people_vaccinated_only_1st_dose':'max',
                 'people_fully_vaccinated':'max',
                 'daily_vaccinations_avgd':'median',
                 '%_people_vaccinated':'max'})
```

```python
con_wise_df
```

```python
con_wise_df
```

```python
con_wise_total_vac = con_wise_df['total_vaccinations'].sort_values(ascending=False).head(20)
con_wise_total_vac
```

```python
con_wise_total_vac.sort_values(ascending=True, inplace=True)
```

## What country have vaccinated more people ?

```python
con_wise_total_vac.plot(kind='barh', figsize=(10,8))

# plt.barh(con_wise_total_vac.index, con_wise_total_vac.values)
plt.ylabel('Conutry',fontsize=14)
plt.xlabel('Total Vaccinations',fontsize=14)
plt.yticks(fontsize=13)
plt.xticks(fontsize=13)
plt.suptitle('Total Vaccinations by Country', fontsize=20)
```

`United States has Vaccinated most number of people`

`USA and China were worst hit by the pandemic and we can see that now they are leading in terms of vaccination administered i.e. people who have atleast received first dose`

`From above bar plot USA, China and India have done mass vaccination`

### Country wise people fully vaccinated

```python
con_wise_full_vac = con_wise_df['people_fully_vaccinated'].sort_values(ascending=False).head(20)
```

```python
help(pd.Series.plot)
```

```python
con_wise_full_vac.plot(kind='bar',figsize=(10,6))
plt.xlabel('Conutry',fontsize=14)
plt.ylabel('Fully Vaccinated',fontsize=14)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.suptitle('People Fully Vaccinated by Country', fontsize=20)
```

`Maximum number of people are given both the doses(1st and 2nd) in United States, followed by India and Turkey`

## How many COVID-19 vaccines doses are administered daily ?

## Which countries have vaccinated more people per day ?

```python
con_wise_daily_vac = con_wise_df['daily_vaccinations_avgd'].sort_values(ascending=False).head(20)
```

```python
con_wise_daily_vac.plot(kind='bar',figsize=(10,6))

plt.xlabel('Conutry',fontsize=14)
plt.ylabel('Median Daily Vaccinations',fontsize=14)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.suptitle('Average Daily Vaccination by Country', fontsize=20)
```

`We can see that USA, China and India and UK has immunized good amount of people on daily basis. Out of which US has highest numbers`

## What country is vaccinated a larger percentage of its population ?

### "What share of the population has received at least one dose of the COVID-19 vaccine?"

### "What share of the population has been fully vaccinated against COVID-19?"

```python
con_max_pop_vac = con_wise_df['%_people_vaccinated'].sort_values(ascending=False).head(20)
```

```python
con_max_pop_vac.plot(kind='bar', figsize=(10,6))
plt.xlabel('Conutry',fontsize=14)
plt.ylabel('Percent(%)',fontsize=14)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.suptitle('Percent of Population Vaccinated by Country', fontsize=20)
```

`From above plot we can see, Although USA has vaccinated much larger people, it has not vaccinated largest percent of population`

`We can say that Gilbratar, Israel and UAE have almost completely vaccinated their country`

```python
con_wise_full_half_vac = con_wise_df.sort_values(by=['total_vaccinations'],
                                                ascending=False)[['people_fully_vaccinated',
                                                                 'people_vaccinated_only_1st_dose']].head(40)
con_wise_full_half_vac
```

```python
con_wise_full_half_vac.plot.bar(stacked=True, figsize=(13,6))
plt.legend(['Fully vaccinated','Atleat one dose'])
# plt.yscale('log')
plt.xlabel('Conutry',fontsize=15)
plt.ylabel('Count',fontsize=14)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.suptitle('Share of Full and half vaccinated', fontsize=20)
```
