## 1.8 Cleaning 'daily_vaccination' column again

```python
prev_vac = 0
prev_con = ''
for idx in range(len(vac_data)):
    curr_con = vac_data.loc[idx, 'country']
    curr_vac = vac_data.loc[idx, 'total_vaccinations']
    
    if(prev_con==''): prev_con = curr_con
    
    
    if(prev_con!=curr_con):
        vac_data.loc[idx, 'daily_vaccinations']=curr_vac
        
    else :
        vac_data.loc[idx, 'daily_vaccinations']=curr_vac - prev_vac
    
    prev_con = curr_con
    prev_vac = curr_vac
```
