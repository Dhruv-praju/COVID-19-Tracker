```python
# Drop all columns that we don't want
vac_data.drop(['iso_code', 'daily_vaccinations_per_million', 'source_name', 'source_website','daily_vaccinations'], axis=1, inplace=True)
```

### Rename columns

```python
vac_data.rename(columns = {'people_vaccinated':'people_only_1st_dose',
                           'people_fully_vaccinated':'people_2nd_dose',
                           'daily_vaccinations_raw':'daily_vaccinations',
                           'total_vaccinations_per_hundred':'%_people_vaccinated',
                           'people_vaccinated_per_hundred':'%_people_only_1st_dose',
                           'people_fully_vaccinated_per_hundred':'%_people_2nd_dose'}, inplace=True)
```

Percentage of missing values

```python
total_mis_val = vac_data.isnull().sum().sum()
total_mis_val
```

```python
total_cells = np.product(vac_data.shape)
total_cells
```

```python
per_missing_val = (total_mis_val / total_cells) * 100
per_missing_val
```
