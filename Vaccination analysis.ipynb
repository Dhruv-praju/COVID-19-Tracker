### 1. 3  Rename columns

```python
vac_data.rename(columns = {'people_vaccinated':'people_vaccinated_only_1st_dose',
                           'daily_vaccinations_raw':'daily_vaccinations',
                           'total_vaccinations_per_hundred':'%_people_vaccinated',
                           'people_vaccinated_per_hundred':'%_people_only_1st_dose',
                           'people_fully_vaccinated_per_hundred':'%_people_fully_vaccinated'}, inplace=True)
```

### 1. 4 Calculate missing values

```python
vac_data.shape
```

```python
vac_data.country.describe()
```

```python
vac_data.isnull().sum()
```

    Percentage of missing values

```python
total_mis_val = vac_data.isnull().sum().sum()
total_mis_val
```

```python
total_cells = np.product(vac_data.shape)
total_cells
```

```python
per_missing_val = (total_mis_val / total_cells) * 100
per_missing_val
```
