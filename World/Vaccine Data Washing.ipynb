### 1.6 Handling missing values

```python
vac_data[vac_data.country=='Wales']
```

```python
vac_data.tail(500)
```

```python
country_grp = vac_data.groupby('country')
country_grp
```

```python
country_grp[['total_vaccinations','people_vaccinated']].fillna(method='ffill').isnull().sum()
```

```python
vac_data['total_vaccinations'].isnull().sum()
```

```python
vac_data['total_vaccinations'] = country_grp['total_vaccinations'].fillna(method='ffill').fillna(0)
```

```python
vac_data['total_vaccinations'].isnull().sum()
```

```python
vac_data
```

```python
vac_data.head(500)
```

```python
vac_data['people_vaccinated'] = country_grp['people_vaccinated'].fillna(method='ffill')
```

```python
vac_data['people_vaccinated'].isnull().sum()
```

```python
vac_data.head(500)
```

```python
vac_data.loc[:,'total_vaccinations':'people_vaccinated'] = vac_data.loc[:,'total_vaccinations':'people_vaccinated'].fillna(axis=1,method='ffill')
```

```python
vac_data.head(100)
```

```python
vac_data['people_vaccinated'].isnull().sum()
```

```python
vac_data['people_fully_vaccinated'] = vac_data['total_vaccinations'] - vac_data['people_vaccinated']
```

```python
vac_data['people_fully_vaccinated'].isnull().sum()
```

```python
vac_data.head(400)
```

```python
vac_data['daily_vaccinations'].fillna(0, inplace=True)
```

```python
vac_data['daily_vaccinations'].isnull().sum()
```

```python
vac_data
```

```python
vac_data['daily_vaccinations_avg'].fillna(0, inplace=True)
```

```python
vac_data.head(200)
```

```python
vac_data['total_vaccinations_per_hundred'] = country_grp['total_vaccinations_per_hundred'].fillna(method='ffill').fillna(0)
```

```python
vac_data['total_vaccinations_per_hundred'].isnull().sum()
```

```python
vac_data.head(200)
```

```python
vac_data['people_vaccinated_per_hundred'] = country_grp['people_vaccinated_per_hundred'].fillna(method='ffill')
```

```python
vac_data['people_vaccinated_per_hundred'].isnull().sum()
```

```python
vac_data.head(200)
```

```python
vac_data.loc[:,'total_vaccinations_per_hundred':'people_vaccinated_per_hundred'] = vac_data.loc[:,'total_vaccinations_per_hundred':'people_vaccinated_per_hundred'].fillna(axis=1,method='ffill')
```

```python
vac_data['people_vaccinated_per_hundred'].isnull().sum()
```

```python
vac_data['people_fully_vaccinated_per_hundred'] = vac_data['total_vaccinations_per_hundred'] - vac_data['people_vaccinated_per_hundred']
```
