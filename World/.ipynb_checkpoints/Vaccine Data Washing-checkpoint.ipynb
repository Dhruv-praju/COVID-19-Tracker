# 1) Data Cleaning

### 1.  2 Drop all columns that we don't want

```python

vac_data.drop(['iso_code', 'daily_vaccinations_per_million', 'source_name', 'source_website'], axis=1, inplace=True)
```

```python
vac_data.info()
```

```python
pd.set_option('display.max_rows', 2000)
vac_data
```

### 1.3   Delete duplicate data

    Countries like Whales, England, Scotland and Northern Ireland come in United Kingdom .
    So delete data of these countries and and use UK data

```python
filt = vac_data.country.isin(['Wales', 'England', 'Scotland', 'Northern Ireland'])
vac_data[filt]
```

```python
vac_data.drop(index=vac_data[filt].index, inplace=True)
```

### 1. 4  Rename columns

```python
vac_data.rename(columns = {'people_vaccinated':'people_vaccinated_only_1st_dose',
                           'daily_vaccinations':'daily_vaccinations_avgd',
                           'daily_vaccinations_raw':'daily_vaccinations',
                           'total_vaccinations_per_hundred':'%_people_vaccinated',
                           'people_vaccinated_per_hundred':'%_people_vaccinated_only_1st_dose',
                           'people_fully_vaccinated_per_hundred':'%_people_fully_vaccinated'}, inplace=True)
```

```python
vac_data
```

```python
vac_data.info()
```

### 1. 5 Calculate missing values

```python
vac_data.shape
```

```python
vac_data.country.describe()
```

```python
vac_data.isnull().sum()
```

    Percentage of missing values

```python
total_mis_val = vac_data.isnull().sum().sum()
total_mis_val
```

```python
total_cells = np.product(vac_data.shape)
total_cells
```

```python
per_missing_val = (total_mis_val / total_cells) * 100
per_missing_val
```

### 1.6 Handling missing values

```python
vac_data[vac_data.country=='Wales']
```

```python
vac_data.tail(500)
```

```python
country_grp = vac_data.groupby('country')
country_grp
```

```python
country_grp[['total_vaccinations','people_vaccinated_only_1st_dose']].fillna(method='ffill').isnull().sum()
```

```python
vac_data['total_vaccinations'].isnull().sum()
```

```python
vac_data['total_vaccinations'] = country_grp['total_vaccinations'].fillna(method='ffill').fillna(0)
```

```python
vac_data['total_vaccinations'].isnull().sum()
```

```python
vac_data
```

```python
vac_data.head(500)
```

```python
vac_data['people_vaccinated_only_1st_dose'] = country_grp['people_vaccinated_only_1st_dose'].fillna(method='ffill')
```

```python
vac_data['people_vaccinated_only_1st_dose'].isnull().sum()
```

```python
vac_data.head(500)
```

```python
vac_data.loc[:,'total_vaccinations':'people_vaccinated_only_1st_dose'] = vac_data.loc[:,'total_vaccinations':'people_vaccinated_only_1st_dose'].fillna(axis=1,method='ffill')
```

```python
vac_data.head(100)
```

```python
vac_data['people_vaccinated_only_1st_dose'].isnull().sum()
```

```python
vac_data['people_fully_vaccinated'] = vac_data['total_vaccinations'] - vac_data['people_vaccinated_only_1st_dose']
```

```python
vac_data['people_fully_vaccinated'].isnull().sum()
```

```python
vac_data.head(400)
```

```python
vac_data['daily_vaccinations'].fillna(0, inplace=True)
```

```python
vac_data['daily_vaccinations'].isnull().sum()
```

```python
vac_data
```

```python
vac_data['daily_vaccinations_avgd'].fillna(0, inplace=True)
```

```python
vac_data.head(200)
```

```python
vac_data['%_people_vaccinated'] = country_grp['%_people_vaccinated'].fillna(method='ffill').fillna(0)
```

```python
vac_data['%_people_vaccinated'].isnull().sum()
```

```python
vac_data.head(200)
```

```python
vac_data['%_people_vaccinated_only_1st_dose'] = country_grp['%_people_vaccinated_only_1st_dose'].fillna(method='ffill')
```

```python
vac_data['%_people_vaccinated_only_1st_dose'].isnull().sum()
```

```python
vac_data.head(200)
```

```python
vac_data.loc[:,'%_people_vaccinated':'%_people_vaccinated_only_1st_dose'] = vac_data.loc[:,'%_people_vaccinated':'%_people_vaccinated_only_1st_dose'].fillna(axis=1,method='ffill')
```

```python
vac_data['%_people_vaccinated_only_1st_dose'].isnull().sum()
```

```python
vac_data['%_people_fully_vaccinated'] = vac_data['%_people_vaccinated'] - vac_data['%_people_vaccinated_only_1st_dose']
```

```python
vac_data['%_people_fully_vaccinated'].isnull().sum()
```

```python
vac_data.head(300)
```

```python
vac_data.isnull().sum()
```

```python
vac_data.info()
```
