## 1.3 Calculate & Handle missing values

```python
covid_df.isnull().sum()
```

```python
filt = covid_df.total_deaths.isnull()
covid_df[filt].country.unique()
```

```python
covid_df['new_deaths'] = covid_df['new_deaths'].fillna(0)
```

```python
covid_df['new_deaths'].isnull().sum()
```

```python
covid_df.tail(200)
```

```python
covid_df['new_cases'] = covid_df['new_cases'].fillna(0)
```

```python
covid_df['new_cases'].isnull().sum()
```

```python
covid_df.tail(1000)
```

```python
country_grp = covid_df.groupby('country')
```

```python
covid_df['total_cases'] = country_grp['total_cases'].transform(lambda x: x.fillna(method='ffill'))
```

```python
covid_df.head(500)
```

```python
covid_df['total_cases'].isnull().sum()
```

```python
covid_df['total_cases'].fillna(0, inplace=True)
```

```python
covid_df['total_deaths'] = country_grp['total_deaths'].transform(lambda x: x.fillna(method='ffill'))
```

```python
covid_df['total_deaths'].isnull().sum()
```

```python
covid_df['total_deaths'].fillna(0, inplace=True)
```

```python
covid_df['total_deaths'].isnull().sum()
```

```python
covid_df.isnull().sum()
```

```python
covid_df.head(300)
```

## Change datatype of columns

```python
covid_df.info()
```

```python
covid_df.loc[:,['new_cases', 'new_deaths', 'total_cases', 'total_deaths']] = covid_df.loc[:,['new_cases', 'new_deaths', 'total_cases', 'total_deaths']].astype('int64')
```
